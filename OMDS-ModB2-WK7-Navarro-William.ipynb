```python
#example
import matplotlib.pyplot as plt
numdates = 12
np.random.seed(0)
time_series1 = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
np.random.seed(1)
time_series2 = 100 * (1 - (np.random.normal(size=numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')

# Plot the first time series
ax = sns.lineplot(x=datearray, y=time_series1, label='Nitrate level')
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")

# Highlight the last three months
sns.lineplot(x=datearray[-3:], y=time_series1[-3:], ax=ax, color='red', linewidth=2)

# Plot the second time series
sns.lineplot(x=datearray, y=time_series2, label='Phosphate level')

# Highlight the last three months
sns.lineplot(x=datearray[-3:], y=time_series2[-3:], ax=ax, color='orange', linewidth=2)

plt.legend()
plt.show()
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize data
numdates = 12
np.random.seed(0)
time_series1 = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
np.random.seed(1)
time_series2 = 100 * (1 - (np.random.normal(size=numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')

# Create the plot
plt.figure(figsize=(12, 6))

# Plot the first time series
sns.lineplot(x=datearray, y=time_series1, label='Nitrate level')
sns.scatterplot(x=datearray[-3:], y=time_series1[-3:], color='red', s=100, zorder=5)

# Plot the second time series
sns.lineplot(x=datearray, y=time_series2, label='Phosphate level')
sns.scatterplot(x=datearray[-3:], y=time_series2[-3:], color='orange', s=100, zorder=5)

# Set labels
plt.xlabel("Year and month")
plt.ylabel("% of eutrophic value")
plt.title("Nitrate and Phosphate Levels Over Time")

# Show legend
plt.legend()

# Display the plot
plt.show()
```
