# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
#Analaysis
#First Approach (Separate Subplots)
#Advantages:
#Allows for clear visual separation between the two weeks
#May be easier to analyze trends within each individual week
#Reduces visual clutter if there are many data points or if the lines would frequently cross
#Enables different y-axis scales for each week if needed (though in your code they share the same scale)

#Disadvantages:
#Makes direct comparison between specific days across weeks more difficult
#Requires the viewer to look back and forth between plots to compare
#Takes up more horizontal space in the visualization
#Setting wspace = 0 removes the gap between subplots which could make axis labels harder to read
#The right subplot is missing y-axis labels in your code (since they're only applied to the left subplot)

#Second Approach (Single Plot with Multiple Lines)
#Advantages:
#Facilitates direct day-to-day comparison between weeks
#Makes overall patterns and differences between weeks immediately visible
#More space-efficient
#Easier to identify specific points where trends diverge between weeks
#Simpler code with fewer commands needed

#Disadvantages:
#Can become visually cluttered if there are many weeks or high variability
#Without additional formatting (like different colors, line styles, or a legend), it might be difficult to distinguish which line represents which week
#Less appropriate if the scales between weeks differ significantly

#the second approach seems to be the best option
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Set a seed for reproducibility
np.random.seed(42)

# Define parameters
years = [2019, 2020, 2021, 2022, 2023]
products = ['Laptop', 'Smartphone', 'Headphones', 'Tablet']

data = {}
for product in products:
    # Create base price between 100 and 1000
    base_price = np.random.randint(100, 1000)
    
    # Create random growth factor between 1.02 and 1.08 (2-8% annual increase)
    growth_factor = 1 + np.random.uniform(0.02, 0.08)
    
    # Generate prices with some randomness
    prices = [round(base_price * (growth_factor ** i) * (1 + np.random.uniform(-0.05, 0.05))) 
              for i in range(len(years))]
    
    data[product] = prices

# Create DataFrame
df = pd.DataFrame(data, index=years)

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Set width of bars
bar_width = 0.2
positions = np.arange(len(years))

# Plot bars for each product
for i, product in enumerate(products):
    offset = (i - len(products)/2 + 0.5) * bar_width
    ax.bar(positions + offset, df[product], bar_width, label=product, alpha=0.8)

# Add labels, title and legend
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Average Price ($)', fontsize=12)
ax.set_title('Average Retail Product Prices by Year', fontsize=16, fontweight='bold')
ax.set_xticks(positions)
ax.set_xticklabels(years)
ax.legend()

# Add grid for better readability
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of each bar
for i, product in enumerate(products):
    offset = (i - len(products)/2 + 0.5) * bar_width
    for j, price in enumerate(df[product]):
        ax.text(j + offset, price + 20, f'${price}', ha='center', va='bottom', fontsize=8)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
